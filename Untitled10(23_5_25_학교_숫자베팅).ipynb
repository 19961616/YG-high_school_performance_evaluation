```python
import random as RR
bonus_question={'Q1':'1번 문제: abs(-2)를 했을 때 출력되는 값을 입력하세요.','Q2':'2번 문제: hex(27)을 했을 때 출력되는 값을 입력하세요.','Q3':'3번 문제: pow(2,-2)를 했을 때 출력되는 값을 입력하세요.'}
bonus_answer={'Q1':str(abs(-2)),'Q2':str(hex(27)),'Q3':str(pow(2,-2))}
def betting(ps,ts,bs,s0):
    s0=1
    if ps>=ts:
        print('점수를 이미 초과하셨습니다.')
        s0=404
        return ps,ts,bs,s0
    while ps<ts and s0==1:
        bs=int(input(f'목표 점수까지 {ts-ps}남았습니다. 보유 점수 {ps} 중에서 베팅할 점수를 입력하세요.: '))
        if 0<bs<=ps:
            ps=ps-bs
            s0=0
            return ps,ts,bs,s0
        elif bs<=0 or bs>ps:
            print('베팅 점수가 0 미만이거나 보유 점수를 초과하는 것은 옳지 않은 입력입니다.다시 베팅하세요.')
def fn_bonus(ps,bq,ba,ts):
    print(f'목표 점수까지 {ts-ps}남았습니다. 현재 보유 점수는 {ps}입니다. 보너스는 주어진 무작위 문제를 맞추면 점수 50을 줍니다.')
    bonus_percentage_0=RR.uniform(0,1)
    if bonus_percentage_0<=0.3:
        print(bq['Q1'])
        answer=str(input('정답을 입력하세요.: '))
        if answer==ba['Q1']:
            ps+=50
            print(f'정답을 맞추셨습니다. 현재 보유 점수는 {ps}이며 목표 점수까지 {ts-ps}남았습니다. 보너스를 나갑니다.')
            return ps
        elif answer!=ba['Q1']:
            print(f'정답을 틀리셨습니다. 현재 보유 점수는 {ps}이며 목표 점수까지 {ts-ps}남았습니다. 보너스를 나갑니다.')
    elif 0.3<bonus_percentage_0<=0.6:
        print(bq['Q2'])
        answer=str(input('정답을 입력하세요.: '))
        if answer==ba['Q2']:
            ps+=50
            print(f'정답을 맞추셨습니다. 현재 보유 점수는 {ps}이며 목표 점수까지 {ts-ps}남았습니다. 보너스를 나갑니다.')
            return ps
        elif answer!=ba['Q2']:
            print(f'정답을 틀리셨습니다. 현재 보유 점수는 {ps}이며 목표 점수까지 {ts-ps}남았습니다. 보너스를 나갑니다.')
    elif 0.6<bonus_percentage_0<=0.9:
        print(bq['Q3'])
        answer=str(input('정답을 입력하세요.: '))
        if answer==ba['Q3']:
            ps+=50
            print(f'정답을 맞추셨습니다. 현재 보유 점수는 {ps}이며 목표 점수까지 {ts-ps}남았습니다. 보너스를 나갑니다.')
            return ps
        elif answer!=ba['Q3']:
            print(f'정답을 틀리셨습니다. 현재 보유 점수는 {ps}이며 목표 점수까지 {ts-ps}남았습니다. 보너스를 나갑니다.')
print('기본 보유 점수는 100으로 시작합니다.')
print('1~5 무작위 자연수를 맞추면 성공하여 점수를 획득하며, 반복하여 목표 점수 이상으로 모으십시오.')
print('보유 점수 내에서 베팅을 하시면, 1,2,3,4,5에 따라 성공시 각각 1,2,3,4,5*점수를 합니다.')
print('자연수 1,2,3,4,5에 따라 5/15, 4/15, 3/15, 2/15, 1/15의 확률로 나옵니다.')
print(' ex) 보유 점수 100에서 베팅 20을 하고 2가 나오면 2배율을 하여 성공하면 정산 점수 140, 실패하면 정산 점수 80이됩니다.')
player_score,betting_score,stop_0=100,0,0
target_score=int(input('101 이상의 목표 점수를 입력하세요.: '))
go_stop=int(input('종료는 0 계속은 1을 입력하세요.: '))
if go_stop==0:
    print('종료하셨습니다.')
while go_stop==1:
    bonus_percentage=RR.random()
    if bonus_percentage<0.5:
        print('특정 확률에 당첨되어 보너스를 받을 기회가 주어졌습니다.')
        player_score=fn_bonus(player_score,bonus_question,bonus_answer,target_score)
    player_score,target_score,betting_score,stop_0=betting(player_score,target_score,betting_score,stop_0)
    if stop_0==404:
        print('점수를 이미 초과하셨습니다. 따라서 종료됩니다.')
        go_stop=0
        break
    Random_number=[1,2,3,4,5]
    random_percentage=[5/15,4/15,3/15,2/15,1/15]
    random_number=RR.choices(Random_number,weights=random_percentage,k=1)[0]
    player_select=int(input('1~5 사이의 예상 숫자를 입력하세요.: '))
    if player_select==random_number:
        if random_number==1:
            player_score=(player_score)+(betting_score)+(betting_score*1)
        elif random_number==2:
            player_score=(player_score)+(betting_score)+(betting_score*2)
        elif random_number==3:
            player_score=(player_score)+(betting_score)+(betting_score*3)
        elif random_number==4:
            player_score=(player_score)+(betting_score)+(betting_score*4)
        elif random_number==5:
            player_score=(player_score)+(betting_score)+(betting_score*5)
    elif player_select!=random_number:
        print(f'1~5의 범위 중에서 {random_number}(이/가) 나왔습니다. 따라서 틀리셨습니다, 정산 결과 현재 보유 점수는 {player_score}입니다. 목표 점수까지 {target_score-player_score}남았습니다.')
    if player_select==random_number:
        print(f'1~5의 범위 중에서 {random_number}(이/가) 나왔습니다. 따라서 맞추셨습니다, 정산 결과 현재 보유 점수는 {player_score}입니다. 목표 점수까지 {target_score-player_score}남았습니다.')
    if player_score<=0:
        print('탕진하셨습니다. 따라서 종료됩니다.')
        break
    if player_score>=target_score:
        print(f'축하합니다 목표 점수인 {target_score} 이상을 모으셨습니다.')
        break
    go_stop=int(input('종료는 0 계속은 1을 입력하세요.: '))
    if go_stop==0:
        print('종료하셨습니다.')
        break
```
